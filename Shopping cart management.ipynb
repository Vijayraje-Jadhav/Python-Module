```python
# 4. Shopping Cart System
# Scenario:
# You are building a shopping cart system where users can add items to the cart, remove items, view the cart, and check out.

# Menu:

# Add Item
# Remove Item
# View Cart
# Checkout
# Exit
# Data Structure: Dictionary to store items (key: item name, value: quantity).
```

```python
def adding(cart):
    item = input("enter the item you want to add: ")
    quantity = int(input("enter the quantity of item: "))
    
    if item in cart:
        cart[item] += quantity
    else:
        cart[item] = quantity
        
    print(f"the item: {item} with the quantity: {quantity} is addeed succesfully in the cart")
    
    
def removing(cart):
    item = input("enter the item you want to remove")
    if item in cart:
        quantity = int(input("enter the quantity you want to remove"))
        
        if quantity >= cart[item]:
            del cart[item]
            print("the item is removed succesfully")
            
        else:
            cart[item]-=quantity
            print("the item is removed succesfully")
            
    else:
        print(f"the item {item} is not found in the cart please enter the valid item")
    

def view(cart):
    if not cart:
        print("the cart is empty")
    else:
        print("your cart contains: ")
        for item, quantity in cart.items():
            print(f"- {item} : {quantity}")

def checkout(cart):
    if not cart:
        print("your cart is empty is empty thats why nothing to checkout")
    else:
        view(cart)
        cart.clear()
        print("thaks for shopping here")

def shopping():
    cart={}
    while True:
        
        
        print("**********welcome to the shpooing cart********************")
        print("press 1 to add item")
        print("press 2 to Remove Item")
        print("press 3 to View Cart")
        print("press 4 to Checkout")
        print("press 0 to Exit")
        
        choice = int(input("enter the choice: "))
        
        if choice==1:
            adding(cart)
        elif choice == 2:
            removing(cart)
        elif choice ==3:
            view(cart)
        elif choice ==4:
            checkout(cart)
        elif choice==0:
            print("The program is exiting")
            break
        else:
            print("the input is invalid please enter the right input")
            
shopping()
```
