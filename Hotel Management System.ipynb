```python
# Function to display all guests
def view_guests(guest_list):
    print("\n--- List of Guests ---")
    if guest_list:
        for guest in guest_list:
            print(f"Guest: {guest[0]}, Room: {guest[1]}, Special Request: {guest[2]}")
    else:
        print("No guests have been added yet.")

# Function to add a new guest
def add_guest(guest_list):
    name = input("Enter guest's name: ")
    room_number = input("Enter room number: ")
    special_request = input("Enter any special request (or 'None' if no requests): ")
    
    guest = (name, room_number, special_request)
    guest_list.append(guest)
    print(f"Guest '{name}' has been added to room {room_number}.")

# Function to search for a guest by room number
def search_guest_by_room(guest_list):
    room_number = input("Enter room number to search for: ")
    found = False
    for guest in guest_list:
        if guest[1] == room_number:
            print(f"Guest found: {guest[0]} is in room {guest[1]} with special request: {guest[2]}")
            found = True
            break
    if not found:
        print(f"No guest found in room {room_number}.")

# Menu-driven function for hotel management system
def hotel_management_menu():
    guest_list = []

    while True:
        print("\n--- Hotel Management System ---")
        print("1. View All Guests")
        print("2. Add New Guest")
        print("3. Search Guest by Room Number")
        print("4. Exit")
        
        choice = input("Enter your choice: ")

        if choice == '1':
            view_guests(guest_list)
        elif choice == '2':
            add_guest(guest_list)
        elif choice == '3':
            search_guest_by_room(guest_list)
        elif choice == '4':
            print("Exiting hotel management system.")
            break
        else:
            print("Invalid choice. Please try again.")

# Call the menu function to start the hotel management system
hotel_management_menu()
```
