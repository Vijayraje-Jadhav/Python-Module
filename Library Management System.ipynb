```python
# 3. Library Book Management
# Scenario:
# You are creating a library book management system. Users can add books to the library, borrow books, return books, and view the list of available books.

# Menu:

# Add a Book
# View Available Books
# Borrow a Book
# Return a Book
# Exit
```

```python
def  add_book(avail_books):
    book_name = input("Enter the book name to add ")
    avail_books.add(book_name)
    print(f"The book name: {book_name} is successfully added in the library")
    
def view_books(avail_books):
    if not avail_books:
        print("the library has no books")
    else:
        print("Available books")
        for i in avail_books:
            print(f"-{i}")
            
            
def borrow_book(avail_books, borrowed_books):
    view_books(avail_books)
    book_title = input("Enter the title of the book to borrow: ")
    if book_title in avail_books:
        avail_books.remove(book_title)
        borrowed_books[book_title] = input("Enter borrower's name: ")
        print(f"Book '{book_title}' borrowed successfully.")
    else:
        print("Book not available.")
        
def return_book(avail_books, borrowed_books):
    
    book_title = input("Enter the title of the book to return: ")
    if not  borrowed_books:
         print("the book you have entered is not borrowed")
        
    else:
        for books, borrower in borrowed_books.items():
            print(f"the {books} borrowed by {borrower}")
        if book_title in borrowed_books:
            
            avail_books.add(book_title)
            del borrowed_books[book_title]
            print("the book is succesfully returned")
        else:
            print("the book was never borrowed ")

        
def book_management():
    avail_books = set()
    borrowed_books = {}
    
    while True:
        
        print("************* welcome labrary management system **************")
        print("press 1 for add a book")
        print("press 2 for view available books")
        print("press 3 for borrow a book")
        print("press 4 for return a book")
        print("press 0 for  exit")
        
        choice = input("Enter your choice: ")

        if choice == '1':
            add_book(avail_books)
        elif choice == '2':
            view_books(avail_books)
        elif choice == '3':
            borrow_book(avail_books, borrowed_books)
        elif choice == '4':
            return_book(avail_books, borrowed_books)
        elif choice == '0':
            print("Exiting library management system.")
            break
        else:
            print("Invalid choice. Please try again.")
    
    
    
book_management()
```
