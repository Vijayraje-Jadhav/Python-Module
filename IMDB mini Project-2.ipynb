```python
# 11. Write a Pandas program to access those movies, released after 1995-01-01.

data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

movies = data[data['release_date']>'1995-01-01']

print(movies.head())
```

```python
# 13. Write a Pandas program to get those movies whose revenue more than 2 million and spent less than 1 million. 

data["revenue"] = pd.to_numeric(data["revenue"], errors = 'coerce')
data["budget"] = pd.to_numeric(data["budget"], errors = 'coerce')

movies = data[(data["revenue"]>2000000) & (data["budget"] <1000000)]

print(movies)
```

```python
# 14. Write a Pandas program to get the longest runtime and shortest runtime.

data["runtime"] = pd.to_numeric(data["runtime"], errors = "coerce")

long_runtime= data["runtime"].max()

short_runtime = data["runtime"].min()

print(long_runtime)
print(short_runtime)
```

```python
# 15. Write a Pandas program to calculate the number of votes garnered by the 70% movie. 

data["vote_count"] = pd.to_numeric(data["vote_count"], errors = "coerce")

percent = data["vote_count"].quantile(0.70)

above= data[data["vote_count"]>=percent]

sum_total = above["vote_count"].sum()

print(sum_total)
```

```python
# # 16. Write a Pandas program to display the movies (title, runtime) longer
# than 30 minutes and shorter than 360 minutes.

data["runtime"] = pd.to_numeric(data["runtime"], errors ="coerce")

duration = data[(data["runtime"] >30 ) & (data["runtime"] < 360)]

#print(duration["title", "runtime"])
print(duration[["title", "runtime"]])
```

```python
# 17. Write a Pandas program to display the movies (title, number of votes) that received specified number of votes. 

n = float(input("enter the number of votes: "))

data["vote_count"] = pd.to_numeric(data["vote_count"], errors="coerce")

votes = data["vote_count"]==n

print(data[votes][["title", "vote_count"]])

data.plot.hist('vote_count')

plt.hist(votes)
```
